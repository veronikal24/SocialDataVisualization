### A taste of HTML

Your website will be built from HTML — the language that web browsers understand. Before we start creating files and pushing them to GitHub, let's play with HTML right here in the notebook.

Jupyter can render HTML directly using `IPython.display.HTML`. This means you can experiment and see results instantly, without leaving the notebook. Try running the cell below (needless to day, this only works if you download the notebook from the `Code` part of the GitHub repo, it won't render in the `.html` version on the course-page):

````python
from IPython.display import HTML

HTML("""
<div style="font-family: 'Helvetica Neue', Arial, sans-serif; max-width: 600px;
            padding: 30px; background: linear-gradient(135deg, #1a1a2e, #16213e);
            border-radius: 12px; color: white;">
    <h1 style="margin-top: 0; font-size: 28px;">SF Crime Analysis</h1>
    <p style="color: #b0b0c0; line-height: 1.6;">
        An interactive exploration of two decades of crime data in San Francisco —
        uncovering patterns across time, space, and category.
    </p>
    <div style="display: flex; gap: 10px; margin-top: 20px;">
        <a href="#" style="padding: 10px 20px; background: #e94560; color: white;
           text-decoration: none; border-radius: 6px;">Hourly patterns</a>
        <a href="#" style="padding: 10px 20px; background: #0f3460; color: white;
           text-decoration: none; border-radius: 6px;">Crime map</a>
    </div>
</div>
""")
````

That's HTML with inline CSS styling — and it already looks like a piece of a modern website. The key things to notice:

* `<div>`, `<h1>`, `<p>`, `<a>` are HTML **elements** — they define the structure (a container, a heading, a paragraph, a link)
* The `style="..."` attributes control how things look — colors, fonts, spacing, layout
* Everything between `<tag>` and `</tag>` is the content of that element

You don't need to become an HTML/CSS expert. Your LLM will write most of it for you. But having a basic feel for what's going on will help you communicate with your LLM and debug things when they don't look right.

## Part 2: Setting Up GitHub Pages

Time to actually build. [GitHub Pages](https://pages.github.com) gives you a free website at `https://USERNAME.github.io`. The way it works is simple:

1. You create a specially-named repository on GitHub.
2. You put HTML files in it.
3. GitHub serves those files to the world as a website.

When you want to update your site, you change the files locally and `git push`. That's it. No server to manage, no hosting fees, no complicated build tools.

### Adding content: from notebooks to web pages

Your site exists but it's just a "Hello World" page. Time to fill it with the visualizations you've been building throughout the course. You have three types to work with:

1. **Matplotlib figures** (Weeks 1–4) — static plots, exported as images
2. **Plotly charts** (Weeks 5–6) — interactive charts with hover, zoom, and clickable legends
3. **Folium maps** (Week 5) — interactive maps with pan, zoom, and data layers

Each exports differently. Here's how to get each type out of your notebooks and onto your website.

**Matplotlib** figures are static, so you export them as image files. Below are some useful commands. You can run those in the notebook where you created a useful matplotlib figure to export it:

````python
# --- Run this in your Week 1-4 notebooks, where you have matplotlib figures ---

import matplotlib.pyplot as plt

# After creating your figure with plt.plot(), plt.bar(), fig, ax = plt.subplots(), etc:
plt.savefig("yearly_crime_trends.png", dpi=150, bbox_inches='tight')
 
# For scalable, print-quality output, use SVG instead:
# plt.savefig("yearly_crime_trends.svg", bbox_inches='tight')

# Then embed in your HTML with:
# <img src="yearly_crime_trends.png" alt="Yearly crime trends in SF" style="max-width: 100%;">
````

**Plotly** charts are interactive, and Plotly can export them as self-contained HTML files that include all the JavaScript needed for the interactivity to work. Once again, here are some helpful commands for exporting:

````python
# --- Run this in your Week 5 or 6 notebook, where 'fig' is your Plotly figure ---

# Option A: Self-contained file (works offline, but can be 3-5 MB)
fig.write_html("crime_hourly_interactive.html")

# Option B: Load Plotly.js from CDN (much smaller file, needs internet to view)
fig.write_html("crime_hourly_interactive.html", include_plotlyjs='cdn')

# Recommend Option B for your website — visitors will have internet,
# and the smaller file means faster page loads.
````

**Folium** maps export the same way, as self-contained HTML:

````python
# --- Run this in your Week 5 notebook, where 'm' is your Folium map ---

m.save("sf_crime_heatmap.html")

# The resulting file contains everything: map tiles, data layers, interactivity.
# Just copy it into your repo.
````

### Putting visualizations on your site

Once you've exported your files, copy them into your `USERNAME.github.io` repository folder. You have two options for displaying them:

**Option 1: Link as separate pages.** Add a link from your `index.html`:

```html
<a href="crime_hourly_interactive.html">Explore hourly crime patterns →</a>
```

Each visualization gets its own full-screen page. Clean and simple.

**Option 2: Embed within a page using an `<iframe>`.** This lets you put a visualization inside a page alongside text and other content:

```html
<iframe src="crime_hourly_interactive.html" width="100%" height="600" frameborder="0"></iframe>
```

This approach is especially useful for building narrative pages with text and visualizations woven together — which is what you'll do in Week 8.

For **static images** (matplotlib), just use an `<img>` tag:

```html
<img src="images/yearly_crime_trends.png" alt="Yearly crime trends" style="max-width: 100%;">
```

**Pro tip: Preview locally.** For a single page, you can just view index.html in browser. But as you page grows in complexity that solution doesn't work. And it gets annoying to push to GitHub and wait for deployment every time you tweak something. So a neat trick is that you can preview your site locally by running this in your terminal from inside your repo folder:

```
python -m http.server 8000
```

Then open `http://localhost:8000` in your browser. Changes show up instantly when you refresh — much faster than the push-and-wait cycle. Use this for rapid iteration, and only push to GitHub when you're happy with the result.

### The art of the prompt

Not all prompts are created equal. Here's the same request at three levels of specificity:

**Level 1 — Vague:**
> *"Make me a website for data visualizations."*

You'll get something generic and forgettable. The LLM has to guess everything about what you want.

**Level 2 — Directional:**
> *"Create an HTML page for a data science portfolio with a dark theme and cards that link to my visualizations."*

Better. The LLM knows the purpose, has a color direction, and a structural element to work with.

**Level 3 — Specific:**
> *"Create a single-page HTML file for a data science portfolio about San Francisco crime analysis. Use a dark background (#1a1a2e) with light text and teal (#16c79a) accents. Include a full-width header with the title 'SF Crime: Two Decades of Data', a two-sentence introduction, and a responsive grid of cards — each card should have a title, a one-line description, and a link. Add subtle hover effects on the cards. All CSS should be in an embedded `<style>` tag. The page should work on mobile."*

Something like this might give you something close to professional quality on the first try. The key is being specific about **structure** (what sections, how they're laid out), **style** (colors, fonts, effects), and **constraints** (single file, embedded CSS, responsive).

<div class="alert alert-block alert-info">
<b>I don't want you to AI generate the structure AND I don't want you to use AI to create the text</b>. You should come up with the right structure for your project, and you should write the text. You don't need to nail it on the first prompt. Start with something reasonable, look at the result, and refine: *"Make the cards wider"*, *"The header is too tall, shrink it"*, *"Change the accent color to coral"*, *"Add more whitespace between sections"*. Each round gets you closer to what you actually want.
